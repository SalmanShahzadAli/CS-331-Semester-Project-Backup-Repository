```python
import os
from dotenv import load_dotenv
import requests

load_dotenv()
GROQ_KEY = os.getenv("REACT_APP_API_KEY")

print("Key loaded:", GROQ_KEY is not None)
```

```python
def ask_groq(message):

    url = "https://api.groq.com/openai/v1/chat/completions"
    
    headers = {
        "Authorization": f"Bearer {GROQ_KEY}",
        "Content-Type": "application/json"
    }

    data = {
        "model": "llama-3.1-8b-instant",   # Free, fast, powerful
        "messages": [
            {"role": "system", "content": "You are HealthMate, a friendly medical assistant who answers with care and clarity."},
            {"role": "user", "content": message}
        ]
    }

    response = requests.post(url, json=data, headers=headers)
    result = response.json()

    return result["choices"][0]["message"]["content"]
```

```python
res = requests.post(
    "https://api.groq.com/openai/v1/chat/completions",
    json={
        "model": "llama-3.1-8b-instant",
        "messages": [{"role": "user", "content": "Hello"}]
    },
    headers={
        "Authorization": f"Bearer {GROQ_KEY}",
        "Content-Type": "application/json"
    }
)

print(res.json())
```

```python
print("Chat with HealthMate (type 'quit' to exit):\n")

while True:
    user = input("You: ")
    print("You: ",user)
    if user.lower() == "quit":
        print("Ending chat. Take care!")
        break
    
    
    bot_reply = ask_groq(user)
    print("Bot:", bot_reply)
```
